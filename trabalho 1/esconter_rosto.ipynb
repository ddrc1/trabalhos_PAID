```python
import face_recognition
import cv2
import matplotlib.pyplot as plt
```

```python
img_rgb = face_recognition.load_image_file("foto3.jpg")
face_locations = face_recognition.face_locations(img_rgb)
```

```python
face_locations
```

```python
if len(face_locations) > 0:
    for face in face_locations:
        (top, right, bottom, left) = face
        sub_img = img_rgb[top: bottom,  left: right]
        
        sub_img = cv2.GaussianBlur(sub_img,(23, 23), cv2.BORDER_DEFAULT)
        img_rgb[top: bottom,  left: right] = sub_img

plt.imshow(img_rgb)
plt.show()
```
