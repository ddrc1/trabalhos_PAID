```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
import scipy
import scipy.signal
from scipy import ndimage
```

```python
img = cv2.cvtColor(cv2.imread("image1.jpg"), cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(img, cmap="gray")
```

```python
i1 = img#cv2.bitwise_not(img)
plt.figure(figsize=(10,10))
plt.imshow(i1, cmap="gray")
```

```python
kernel_borders = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
i2 = cv2.filter2D(i1, -1, kernel_borders)
plt.figure(figsize=(10,10))
plt.imshow(i2, cmap="gray")
```

```python
i3 = cv2.bitwise_not(i2)
plt.figure(figsize=(10,10))
plt.imshow(i3, cmap="gray")
```
