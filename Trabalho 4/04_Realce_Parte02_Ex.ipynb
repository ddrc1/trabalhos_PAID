## Imports

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
import scipy
import scipy.signal
from scipy import ndimage
```

## Carregando Imagem em Preto e Branco

```python
img = cv2.cvtColor(cv2.imread("lenna.png"), cv2.COLOR_BGR2GRAY)
```

## Apresentando imagem

```python
plt.imshow(img, cmap="gray")
plt.show()
```

## Convolução e Correlação

```python
def filter2d(image, kernel, convolve = False):

    col, row = image.shape

    if convolve:
        kernel = np.rot90(kernel, 2)
        
    new_image = np.zeros(image.shape)

    k_col, k_row = kernel.shape
    center_index = int(np.floor(k_col/2))
    for i in range(1, col-(2*center_index)):
        for j in range(1, row-(2*center_index)):
            new_image[i][j] = np.sum(image[i:i+k_col, j:j+k_row] * kernel)  
    return new_image

kernel = np.array([
                   [-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1],
                ])

print("Imagem resultante da função: ")
result = filter2d(img, kernel, convolve=False)
plt.imshow(result, cmap="gray")
plt.show()

print("Imagem resultante da correlação: ")
result = scipy.signal.correlate2d(img,kernel)
plt.imshow(result, cmap="gray")
plt.show()

print("Imagem resultante da convolução: ")
result = scipy.signal.convolve2d(img,kernel)
plt.imshow(result, cmap="gray")
plt.show()
```
